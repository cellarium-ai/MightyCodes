## MightyCodes `sa-bac` interactive

```python
import numpy as np
import torch
import matplotlib.pylab as plt

from mighty_codes.sa_bac.main import \
    SimulatedAnnealingBinaryAsymmetricChannel

from mighty_codes.plot_utils import \
    plot_sa_trajectory, \
    plot_sa_codebook, \
    plot_sa_resampling_buffer
```

```python
sabac = SimulatedAnnealingBinaryAsymmetricChannel.init_from_yaml(
    input_yaml_file='./sa_bac_params.yaml',
    logger=print)
```

```python
sabac.run()
```

```python
plot_sa_trajectory(
    ptpsa=sabac.ptpsa,
    figsize=(12, 10))
```

```python
plot_sa_codebook(
    ptpsa=sabac.ptpsa,
    codebook_bx=sabac.ptpsa.top_state_kx,
    energy_b=sabac.ptpsa.top_energy_k,
    idx=0,
    batch_name="Top States",
    figsize=(10, 10))
```

```python
plot_sa_resampling_buffer(
    ptpsa=sabac.ptpsa,
    figsize=(10, 4))
```
